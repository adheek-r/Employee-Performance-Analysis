### Function to remove Outlliers

```python
def outliers(df,column):
    Q1 = np.percentile(df[column], 25)
    Q3 = np.percentile(df[column], 75)
    IQR = Q3 - Q1
    LB = Q1 - (1.5 * IQR)
    UB = Q3 + (1.5 * IQR)

    mean_value = df[column].mean()

    df[column] = np.where((df[column] < LB) | (df[column] > UB), mean_value, df[column])

    return df
```

* From the above plot it is clear that all the categories of the column have almost similar average performance rating of the employee.
* So we perform Chi-square to evaulate the any significant association between the categorical variable and target variable.

### Chi-square Method

```python
from scipy.stats import chi2_contingency

for i in df_cat.columns:

    contingency_table = pd.crosstab(df_cat[i], df['Per Rating'])

    _,p,*_= chi2_contingency(contingency_table)

    alpha = 0.05
    if p < alpha:
        print(f'{i} is dependent (significant association).')
    else:
        print(f'{i} is independent (no significant association).')
```

```python
df_cat.drop(['Gender','Maratial','BussTravelFreq','Attrition'],axis=1,inplace=True)
```

* Dropping the columns(Gender, Maratial Status, Business Travell Frequency, Attrition) based on the p-value of Chi-square test and above barplot analysis.

```python
contingency_table = pd.crosstab(df_cat['Department'], df_cat['JobRole'])
_,p,*_= chi2_contingency(contingency_table)
print(p)
```

```python
df_cat.drop(['Department'],axis=1,inplace=True)
```

* Dropping the employee department column because, the employee jobrole column is derived from department column itself and the p-value between those columns are equal, which indicates strong relationship between those columns.

### Function for Encoder

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
def encoder(df,column):
    onehot = OneHotEncoder(sparse_output=False)
    column_encoded = onehot.fit_transform(df[[column]])
    column_encoded_df = pd.DataFrame(column_encoded, columns=onehot.get_feature_names_out([column]))
    column_encoded_df.drop(column_encoded_df.columns[0], axis=1, inplace=True)
    df = pd.concat([df, column_encoded_df], axis=1)
    df.drop([column], axis=1, inplace=True)
    return df
```

```python
model_params = {
    'random_forest': {
        'model': RandomForestClassifier(n_jobs=-1),
        'params': {
            'randomforestclassifier__n_estimators': [1, 5, 10, 100, 200],
            'randomforestclassifier__criterion': ['gini', 'entropy'],
            'randomforestclassifier__min_samples_split': [2, 3]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(multi_class='auto', n_jobs=-1),
        'params': {
            'logisticregression__C': [1, 5, 10, 50, 100],
            'logisticregression__solver': ['lbfgs', 'sag', 'newton-cg', 'liblinear'],
            'logisticregression__max_iter': [100, 200, 300]
        }
    },
    'gbm': {
        'model': GradientBoostingClassifier(loss='log_loss'),
        'params': {
            'gradientboostingclassifier__n_estimators': [1, 5, 10, 100, 200],
            'gradientboostingclassifier__max_depth': [3, 4],
            'gradientboostingclassifier__max_features': ['auto', 'log2']
        }
    },
    'knn': {
        'model': KNeighborsClassifier(),
        'params': {
            'kneighborsclassifier__n_neighbors': [3, 5, 7],
            'kneighborsclassifier__weights': ['uniform', 'distance'],
            'kneighborsclassifier__metric': ['minkowski', 'euclidean', 'manhattan']
        }
    },
    'svm': {
        'model': SVC(),
        'params': {
            'svc__C': [0.1, 1, 10, 100],
            'svc__kernel': ['linear', 'poly', 'rbf'],
            'svc__degree': [2, 3, 4],
            'svc__gamma': ['scale', 'auto'],
            'svc__probability': [True]
        }
    }
}
```

```python
scores=[]
best_estimators={}
for algo,mp in model_params.items():
    pipe=make_pipeline(mp['model'])
    clf=GridSearchCV(pipe,mp['params'],cv=3,return_train_score=False,verbose=3)
    clf.fit(x_sm,y_sm)
    scores.append({
        'model':algo,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
        })
    best_estimators[algo]=clf.best_estimator_

scores
```

```python
scores=pd.DataFrame(scores,columns=['model','best_score','best_params'])
scores
```
